```python
import pandas as pd
import numpy as np
import scipy.stats 
import streamlit as st
import matplotlib 
from matplotlib import pyplot as plt
import altair as alt
import plotly_express as px
```

```python
df = pd.read_csv('vehicles_us.csv')
df
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['model_year'], inplace=True)
df
```

```python
df['model_year']=df['model_year'].astype(int)
df
```

```python
df['cylinders'] = pd.to_numeric(df['cylinders'], errors='coerce')
df.dropna(subset=['cylinders'], inplace=True)
df.sample(10)
```

```python
df.dropna(subset=['odometer'], inplace=True)
df['odometer']=df['odometer'].astype(int)
df
```

```python
df['paint_color']=df['paint_color'].fillna('missing')
df
```

```python
df['is_4wd']=df['is_4wd'].fillna('no')
df['is_4wd'] = df['is_4wd'].replace(1.0, 'yes')
df
```

```python
df['date_posted'] = pd.to_datetime(df['date_posted'])
df['date_posted'] = df['date_posted'].dt.strftime('%m/%d/%Y')
df
```

```python
df['price']=df['price'].astype(float)
df
```

```python
df['manufacturer']=df['model'].apply(lambda x:x.split()[0])
df
```

```python
df.isna().sum()
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
color_map = {
    'new': 'green',
    'like new': 'blue',
    'excellent': 'purple',
    'good': 'orange',
    'fair': 'red',
    'poor': 'grey'
}


fig_hist = px.histogram(
    df, 
    x='price', 
    color='condition', 
    title='Price Distribution by Condition', 
    barmode='overlay',
    color_discrete_map=color_map
)

fig_hist.update_xaxes(range=[0, 85000])

fig_hist.show()
```

```python
color_map = {
    'SUV': 'blue',
    'sedan': 'green',
    'pickup': 'red',
    'truck': 'purple',
    'coupe': 'orange',
    'hatchback': 'pink',
    'van': 'yellow',
    'wagon': 'brown',
    'mini-van': 'gold',
    'convertible': 'silver',
    'other': 'black',
    'bus': 'teal',
    'offroad': 'lime green'

}

fig_scatter = px.scatter(
    df, 
    x='type', 
    y='price', 
    color='type',
    title='Price vs. Type',
    color_discrete_map=color_map
)

fig_scatter.show()
```
